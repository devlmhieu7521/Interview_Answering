```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nyc_subway.csv')
df['conds'].value_counts().plot.bar()
plt.ylabel('quantity of people')
plt.title('The number of people take subway based on weather');
```

Theo như biểu đồ ở trên, việc đi subway của mọi người vào nhiều ngày thời tiết khác nhau. Tuy nhiên, việc mọi người đi subway ở NYC chủ yếu là vào ngày không có mưa và sương mù

```python
df['rain'].value_counts()
```

```python
df = df.groupby(['hour','rain'])['rain'].count().to_frame().unstack()
```

```python
df.plot.bar()
plt.ylabel('quantity of people')
plt.title('The number of people take subway in rainy condition');
```

Ta có thể thấy là vào mỗi khung giờ và điều kiện thời tiết mưa hay không mưa, thì việc số người tham gia subway vào ngày không mưa thường gấp 5 lần vào ngày mưa
